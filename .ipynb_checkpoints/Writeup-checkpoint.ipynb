1. Unigram and Bigram Probabilities: 

 Unigram and bigram probabilities were calculated by tokenizing each sentence/line of the training file, storing the frequencies of each vocabulary in the file and each word with frequency 1 is replaced with "\<unk>". Then, by iterating through tokens, the unigram or bigrams are stored in a list, as well as n-gram frequency dictionaries with keys as a tuple containing the n-grams and values as the frequency of those n-grams. Additionally, n-1 grams along with their frequencies were also stored in dictionaries. Finally in the score function, for a given sentence, we iterate through each token, create n-grams as well as n-1 grams, find those n-grams and n-1 grams in our previously mentioned dictionary to get n-count and n-1-count, and the probability of n-gram can be calculated as n-count/n-1-count. **For convinence, log(n-count + 1/n-1count + |V|) was used.**
         
2. The advantage of using "unk" is to smooth out the probabilities, i.e., delete rarely used words that in aggregate would cause the model less effectively and accurately. However, in the language model in HW2, using <unk> results in a corpus with all words unkown to have lower perplexity than one with some words unkown. For an example of this, in the current language model, unknowns.txt has lower perplexity than iamsam2.txt, which should not be the case. If frequency of <unk> changes from 1 to 3, less words would be recognized by our model and the output random sentences/text files would have less diversity of words. 

3. Using berp-training.txt as training file and hw2-test as our test file, unigrams perform much more poorly than bigrams and trigrams. Both bigrams and trigrams have similar perplexities with the perplexity(2-gram) being 9.14 and perplexity(3-gram) 9.8. Switching to hw2_test-trigram.txt as the test file, trigrams (28) performed better than bigrams (33) but unigrams (21) performed better than both.
